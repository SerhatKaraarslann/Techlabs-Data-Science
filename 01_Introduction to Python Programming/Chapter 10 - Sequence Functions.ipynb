## Python Enumerate Function - Python Quick Tips

```python
# START YOUR CODE HERE.
my_list = ['apples','pears','orange','fruits']

# Way 1
count = 1
for element in my_list:
    print(element)
    if count == 0:
        print("Count is 1")
    count+=1

#way 2
for x in range(len(my_list)):
    print(my_list[x])
    if x == 1:
        print('X is 1')

# enumerate is more elegant way to print index and the element, helps you with loops that require a counter by adding an index to each item in an iterable. 
for x,element in enumerate(my_list):
    print(x,element)
```

## Sorting in Python || Learn Python Programming (Computer Science)

```python
# START YOUR CODE HERE.
earth_metals = ['Beryllium','Magnesium','Calcium','Strontium','Barium','Radium'] # sorted by atomic number

# sorting alphabeticly
earth_metals.sort()
print(earth_metals)

#reverse order
earth_metals.sort(reverse=True)
print(earth_metals)

#tuple sorting
earth_metals = ('Beryllium','Magnesium','Calcium','Strontium','Barium','Radium')
#earth_metals.sort() # AttributeError : 'tuple object has no attribute 'sort'
#tuples are immutable, they can not be changed

#list.sort method
help(list.sort)
#sort(key=None, reverse=True) , key is the sort function
#reverse = False -> sort in ascending order
#reverse = True -> sort in descending order

planets = [('Mercury',2440,5.43,0.395), # Name, #radius at equador,density in g/cmÂ³, distance from sun
           ('Venus',6052,5.24,0.723),
           ('Earth',6378,5.52,1.000),
           ('Mars',3396,3.93,1.530),
           ('Jupiter',71492,1.33,5.210),
           ('Saturn',60268,0.69,9.551),
           ('Uranus',25559,1.27,19.213),
           ('Neptune',24764,1.64,30.070)]

size = lambda planet: planet[1]
planets.sort(key=size,reverse= True) # key is the sort function
print(planets)

density = lambda planet : planet[2]
planets.sort(key = density)
print(planets)

#list.sort changes the list

#if we wanna sort a tuple we have to use sorted() function, this function can be used for the lists too
#sorted() method is the iterable data structure, list, string, tuple,sets, dict, frozensets
earth_metals = ['Beryllium','Magnesium','Calcium','Strontium','Barium','Radium']
sorted_earth_metals = sorted(earth_metals)
print(sorted_earth_metals)

print(earth_metals) # original list is the same, not changed

#tuple sorting
data = (7,2,5,6,1,3,9,10,4,8)
sorted(data) # output of this method is a list and original tuple unaltered


sorted('Alphabetical')
```

## Zip Function - Python Quick Tips

```python
# START YOUR CODE HERE.
x = [1,2,3,4]
y = [1,1,3,7,9]

for i in range(min(len(x),len(y))):
    if x[i] == y [i]:
        print('Same')
    else:
        print('Not same')

# zip function takes any amount of iterable objects and packs them together into tuples
print(list(zip(x,y))) #  (1, 1), (2, 1), (3, 3), (4, 7)]  9 has been ignored

z = ['one','two']
print(list(zip(x,y,z))) # [(1, 1, 'one'), (2, 1, 'two')]

for i,j in zip(x,y):
    if i == j:
        print('Same')
    else:
        print('Not same')
```

#### 2. Write a program to read through the data/mbox-short.txt and figure out emails as well as their line indexes. Lines that contain emails start with "From ". Store the pairs in a list of tuples named *pairs* that have the format (index, email). Sort this list with regard to the emails in alphabetical order.
<br /> 
<details>
<summary>
    <font size="3" color="red"><b>Hints (click to expand)</b></font>
</summary>
<p>
    <ul>
        <li>First, store each email with their regarding index in a list as a tuple.</li>
        <li>Second, sort this list with regard to the emails in alphabetical order.</li>
    </ul>
</p>
</details>

```python
pairs = []
name = input('Enter a file name: ')
if len(name) < 1:
    name = 'data/mbox-short.txt'

with open(name) as f:
    file = f.readlines()
    for index,line in enumerate(file):
        if not line.startswith('From '): continue
        email = line.split()[1]
        pairs.append((index,email))

pairs.sort(key= lambda x:x[1])
```

```python
# THIS CELL TESTS YOUR RESULTS.
assert len(pairs) == 27, "It seems as you did not filter out every pair!"
assert pairs[2][0] == 493, "You filtered out a wrong pair!"
assert pairs[-1][0] == 904, "Your filtered out a wrong pair!"
```
