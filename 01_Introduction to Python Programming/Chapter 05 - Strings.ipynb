## Strings - Part 2

```python
# START YOUR CODE HERE.
s = 'Monty Python'
print(s[0:4])
print(s[6:7])
print(s[6:20])
print(s[:2])
print(s[8:])
print(s[:])

a = 'Hallo'
b= a +'There'
print(b)
c = a+' '+'There'
print(c)

fruit = 'banana'
'n' in fruit
'm' in fruit
if 'a' in fruit:
    print('Found it!')

word = input('Enter a word: ')

if word == 'banana':
    print('All right, bananas.')

if word < 'banana':
    print('Your word,'+word+', comes before banana.')
elif word > 'banana':
    print('Your word,'+word+', comes after banana')
else:
    print('All right, bananas.')

greet = 'Hello Serhat'
zap = greet.lower()
print(zap)
print('Hi There'.lower())

stuff = 'Hello World'
type(stuff)
dir(stuff)
```

```python
#find function
fruit = 'banana'
pos = fruit.find('na')
print(pos)

aa = fruit.find('z')
print(aa)

#upper & lower
greet = 'Hello Serhat'
nnn = greet.upper()
print(nnn)

www = greet.lower()
print(www)

#replace
nstr = greet.replace('Serhat','John')
print(nstr)

nstr = greet.replace('o','X')
print(nstr)

#whitespace
greet = '            Hello Serhat      '
print(greet.lstrip())

print(greet.rstrip())

print(greet.strip())

#Prefixes
line = 'Please have a nice day'
line.startswith('Please')
line.startswith('p')

data = 'From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008'
atpos = data.find('@')
print(atpos)

sppos = data.find(' ',atpos)
print(sppos)

host = data[atpos+1 : sppos]
print(host)
```
