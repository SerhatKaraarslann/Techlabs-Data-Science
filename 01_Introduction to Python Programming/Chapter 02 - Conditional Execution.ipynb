Pay the hourly rate for the hours up to 40 and 1.5 times the hourly rate for all hours worked above 40 hours. Use 45 hours and a rate of 10.50 per hour to test the program (the pay should be 498.75). You should use input to read a string and float() to convert the string to a number. Do not worry about error checking the user input - assume the user types numbers properly. Print your computed pay.

```python
# START YOUR CODE HERE.
hrs = input("Enter Hours:")
h = float(hrs)

rate = input('Enter rate:')
rate = float(rate)


if h > 40:
    pay = 40*rate +(h-40)*(rate*1.5)
else:
    pay = h*rate
    
print(pay)
```

#### 2. Write a program to prompt for a score between 0.0 and 1.0. If the score is out of range, print an error. If the score is between 0.0 and 1.0, print a grade using the following scheme: >= 0.9 A; >= 0.8 B; >= 0.7 C; >= 0.6 D; < 0.6 F. For the test, enter a score of 0.85.

```python
# START YOUR CODE HERE.
score = input("Enter Score: ")

try:
    score = float(score)
    
    if score < 0.0 or score > 1.0:
        exit()
    
    if score >= 0.9:
       print('A')
    elif score >= 0.8:
       print('B')
    elif score >= 0.7:
       print('C')
    elif score >= 0.6:
       print('D')
    else:
       print('F')
except:
    print('The score must be between 0.0 and 1.0')
```
