## Sets - Set Methods and Operations to Solve Common Problems

```python
# sets are some of list, which removes all of duplicate values
# list of values in set function 
s1 = set([1,2,3,4,5])
print(s1)
print(type(s1))

# or with curly brackets { }
s2 = {1,2,3,4,5,1,2,3}
print(s2)
print(type(s2))

# !!!! to create empty set, we have to use set() function, otherwise the curly brackets create an empty dictionary !!!!!

# we can add, update and remove values 
s1 = {1,2,3,4,5}
print(s1)
s1.add(6)
print(s1)


s1.update([6,7,8])
print(s1)

s2 = {7,8,9}
s1.update([6,7,8],s2)
print(s1)

s1.remove(5)
print(s1)

#remove vs. discard
# if i wanna  remove an non exist value in set and use remove method, i met with the key error but discard show me no error and run properly
#s1.remove(10) # KeyError
print(s1)

s1.discard(11)
print(s1)
```

```python
#intersection 
s1 = {1,2,3}
s2 = {2,3,4}
s3 = {3,4,5}

s4 = s1.intersection(s2) #gives me all values that are in s2 and also that are in s1
print(s4) # 2,3

s5 = s1.intersection(s2,s3)
print(s5) # 3

# difference 
# this function makes reverse, what intersection function makes

s6 = s1.difference(s2) # gives me the different values that are in s1 but not in s2
print(s6) # 1

s7 = s2.difference(s1,s3)
print(s7) # empty list

# if i want to see all of variables that are different in both sets, that are either in s1 and not in s2 or in s2 but not in s1
s8 = s1.symmetric_difference(s2)
print(s8) #1,4
```

```python
# cast list in set, remove duplicate values
l1 = [1,2,3,1,2,3]
print(l1)
l2 = set(l1)
print(l2)

l3 = list(set(l1))
print(l3)

employees = ['Corey','Jim','Steven','April','Judy','Jenn','John','Jane']
gym_members = ['April','John','Corey']
developers = ['Judy','Corey','Steven','Jane','April']

result = set(gym_members).intersection(developers) # the employees who has gym memberships and are developers
print(result) 

result = set(employees).difference(gym_members,developers) # all of the employees who are neither gym members nor developers
print(result)

if 'Corey'in developers: #O(n) for list, O(1) for a set
    print('Found!')
```

#### 1. Write a program to read through the data/mbox-short.txt and figure out a non-repeating collection of emails. You can pull the emails from lines that start with "From ".

```python
name = input('Enter a file name: ')
if len(name) < 1:
    name = 'data/mbox-short.txt'

emails = set()
fhand = open(name)
for line in fhand:
    if not (line.startswith('From ')) : continue
    words = line.split()
    email = words[1]
    emails.add(email)

print(emails)
```
